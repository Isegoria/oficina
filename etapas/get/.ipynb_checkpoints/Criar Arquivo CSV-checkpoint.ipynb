# Programa escrito para criar o arquivo csv usado na oficina com os dados do Bolsa Família do Distrito Federal

#### Primeiro definir a função de requisição de dados JSON da Api de Dados do Portal da Transparência

```python
from urllib.request import Request, urlopen
import numpy as np
import pandas as pd
import datetime
import json
import math

siteApi = 'http://www.transparencia.gov.br/api-de-dados'

def getJsonData(strMesAno,strCodIBGE,numPag):
    method='bolsa-familia-por-municipio?mesAno={0}&codigoIbge={1}&pagina={2}'.format(strMesAno,strCodIBGE,numPag)
    req = Request(siteApi + '/' + method, headers={'User-Agent': 'Mozilla/5.0'})
    JSON = urlopen(req).read()
    return json.loads(JSON)
```

#### Agora, fazer diversas requisições, uma para cada mês dentro de um intervalo de datas

```python
import datetime
from dateutil.relativedelta import *

dtDe = datetime.date.fromisoformat('2010-01-01')
dtPara = datetime.date.fromisoformat('2020-12-01')

col_id = []
col_dataReferencia = []
col_codIBGE = []
col_nomeIBGE = []
col_UFSigla = []
col_UFNome = []
col_valor = []
col_qtdBeneficiados = []

dtProx = dtDe
while dtProx <= dtPara:
    # Fabrica Mes no formato YYYYMM
    strMes = dtProx.strftime("%Y%m")
    dtProx = dtProx + relativedelta(months=+1)
    
    # Faz requisição json de Brasilia
    jsonBrasilia = getJsonData(strMes,'5300108',1)
    
    if len(jsonBrasilia) != 0:
        # Adiciona os dados em cada uma das listas
        col_id.append(jsonBrasilia[0]['id'])
        col_dataReferencia.append(pd.to_datetime(jsonBrasilia[0]['dataReferencia']))
        col_codIBGE.append(jsonBrasilia[0]['municipio']['codigoIBGE'])
        col_nomeIBGE.append(jsonBrasilia[0]['municipio']['nomeIBGE'])
        col_UFSigla.append(jsonBrasilia[0]['municipio']['uf']['sigla'])
        col_UFNome.append(jsonBrasilia[0]['municipio']['uf']['nome'])
        col_valor.append(jsonBrasilia[0]['valor'])
        col_qtdBeneficiados.append(jsonBrasilia[0]['quantidadeBeneficiados'])
        print(jsonBrasilia[0]['dataReferencia'])
        
```

#### Criamos um dicionário com as listas geradas

```python
dictBF = {
'id':col_id,
'dataReferencia':col_dataReferencia,
'codIBGE':col_codIBGE,
'nomeIBGE':col_nomeIBGE,
'UFSigla':col_UFSigla,
'UFNome':col_UFNome,
'valor':col_valor,
'qtdBeneficiados':col_qtdBeneficiados,
}
```

#### Criamos o DataFrame pandas

```python
df = pd.DataFrame(data=dictBF)
```

```python
df.to_csv('BolsaFamiliaNovo.csv', index=False)
```
